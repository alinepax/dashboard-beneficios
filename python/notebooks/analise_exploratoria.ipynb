```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo dos gr√°ficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Carregar os dados
colab_df = pd.read_csv('colaboradores_anonimizados.csv')
benef_df = pd.read_csv('uso_beneficios_anonimizados.csv')

# Visualizar os dados
colab_df.head()
```

```python
# Resumo estat√≠stico e tipos
colab_df.info()
benef_df.info()

colab_df.describe()
benef_df['Benef√≠cio_Utilizado'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar gr√°fico
ax = sns.countplot(data=colab_df, x='Departamento', palette='Set2')
plt.title('Quantidade de colaboradores por departamento')
plt.xticks(rotation=45)

# Adicionar n√∫meros inteiros nas barras
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', 
                va='bottom')

# Caminho e nome do arquivo
caminho = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 7 - Sistema de Apoio √† √Årea de Benef√≠cios\graficos'
plt.savefig(f'{caminho}\\colaboradores_por_departamento.png', bbox_inches='tight', dpi=300)

plt.show()
```

### Um Olhar R√°pido na Nossa Equipe! üëÄ

Analisando nosso gr√°fico, temos um insight super legal sobre a distribui√ß√£o dos times:

* **Equil√≠brio √© tudo!** ‚öñÔ∏è Nossos departamentos de `TI`, `Comercial`, `Administrativo` e `Financeiro` s√£o quase do mesmo tamanho. Isso significa que podemos lan√ßar um benef√≠cio novo e o impacto ser√° sentido de forma bem parecida por quase toda a galera!

* **RH: Nosso laborat√≥rio de ideias!** üß™ Sendo nosso menor time, o `RH` √© o grupo perfeito para testar em primeira m√£o aquele benef√≠cio inovador que estamos planejando. Um piloto de sucesso com eles √© o primeiro passo para um lan√ßamento incr√≠vel para toda a empresa! üöÄ‚ú®

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Ordenar os benef√≠cios por uso
beneficio_ordenado = benef_df['Benef√≠cio_Utilizado'].value_counts().index

# Gr√°fico de barras horizontais
ax = sns.countplot(data=benef_df,
                   y='Benef√≠cio_Utilizado',
                   order=beneficio_ordenado,
                   palette='Set2')

plt.title('Benef√≠cios mais utilizados')
plt.xlabel('Quantidade de ades√µes')
plt.ylabel('Benef√≠cio')

# Adicionar n√∫meros nas barras
for p in ax.patches:
    width = p.get_width()
    ax.annotate(f'{int(width)}',
                (width + 0.5, p.get_y() + p.get_height() / 2),
                ha='left',
                va='center')

# Salvar gr√°fico
caminho = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 7 - Sistema de Apoio √† √Årea de Benef√≠cios\graficos'
plt.savefig(f'{caminho}\\beneficios_mais_utilizados.png', bbox_inches='tight', dpi=300)

plt.show()
```

### Nossos Benef√≠cios em A√ß√£o! üöÄ

Este gr√°fico mostra quais benef√≠cios s√£o os mais queridinhos da equipe, e os resultados s√£o muito interessantes!

* **Sa√∫de Mental e F√≠sica no Topo!** üß†üí™

* **O B√°sico que (Sempre) Funciona!** üç≤ü•á
    O `Vale Refei√ß√£o` continua sendo o campe√£o de ades√µes, com 59 pessoas. Ele √© a base s√≥lida do nosso pacote de benef√≠cios e prova seu valor no dia a dia.

* **Ponto de Aten√ß√£o: Plano de Sa√∫de** üßêüîç
    √â curioso que o `Plano de Sa√∫de` tenha uma ades√£o um pouco menor (49 pessoas) em compara√ß√£o aos outros aux√≠lios de bem-estar. Vale a pena investigar o motivo!
```python
import seaborn as sns
import matplotlib.pyplot as plt

# Unir as tabelas
merged = benef_df.merge(colab_df, on='ID_Colaborador')

# Tabela cruzada
tabela = pd.crosstab(merged['Departamento'], merged['Benef√≠cio_Utilizado'])

# Criar o heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(tabela, annot=True, fmt='d', cmap='Oranges')
plt.title('Uso de Benef√≠cios por Departamento')
plt.xlabel('Benef√≠cio')
plt.ylabel('Departamento')
plt.tight_layout()

# Salvar gr√°fico
caminho = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 7 - Sistema de Apoio √† √Årea de Benef√≠cios\graficos'
plt.savefig(f'{caminho}\\uso_beneficios_departamento.png', bbox_inches='tight', dpi=300)

plt.show()
```

### A Jornada do Colaborador e Seus Benef√≠cios! üó∫Ô∏è

Este mapa mostra que a rela√ß√£o do time com os benef√≠cios se transforma ao longo do tempo na empresa. Cada fase tem uma prioridade diferente!

* **Veteranos (> 5 anos): A Base S√≥lida!** üèÜüõ°Ô∏è
    Nossos colaboradores mais antigos s√£o a prova de que o tradicional tem muito valor. Eles s√£o os maiores usu√°rios do `Vale Refei√ß√£o` e do `Plano de Sa√∫de`. * **Alerta de Engajamento no 2¬∫ Ano!** üìâ‚ùì
    Notamos uma queda expressiva no uso de benef√≠cios no grupo de `1 a 2 anos` de casa. Este √© um ponto de aten√ß√£o cr√≠tico! Ser√° um reflexo da "crise do segundo ano", um fen√¥meno conhecido de queda de engajamento ap√≥s o per√≠odo inicial de euforia? √â o momento perfeito para lan√ßar uma campanha de reengajamento, refor√ßando o valor e as possibilidades do nosso pacote de benef√≠cios.

```python
# Garantir que Data_Ades√£o est√° como datetime
benef_df['Data_Ades√£o'] = pd.to_datetime(benef_df['Data_Ades√£o'])
```

```python
# Agrupar por m√™s
benef_por_mes = benef_df.groupby(benef_df['Data_Ades√£o'].dt.to_period('M')).size().reset_index(name='Quantidade')
benef_por_mes['Data_Ades√£o'] = benef_por_mes['Data_Ades√£o'].dt.to_timestamp()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.lineplot(data=benef_por_mes, x='Data_Ades√£o', y='Quantidade', marker='o', color='#FF6F00')
plt.title('Evolu√ß√£o mensal de ades√£o aos benef√≠cios')
plt.xlabel('M√™s')
plt.ylabel('Quantidade de ades√µes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Salvar o gr√°fico
caminho = r'C:\Users\Aline.Aline\Desktop\GitHub\Projeto 7 - Sistema de Apoio √† √Årea de Benef√≠cios\graficos'
plt.savefig(f'{caminho}\\evolucao_mensal_beneficios.png', bbox_inches='tight', dpi=300)
plt.show()
```
